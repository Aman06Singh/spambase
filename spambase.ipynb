# 1. Loading Dataset

```python
import pandas as pd
```

```python
def load_spam_data(fileName):

    input_columns = ['word_freq_make', 'word_freq_address', 'word_freq_all', 'word_freq_3d', 'word_freq_our', 'word_freq_over', 'word_freq_remove',
               'word_freq_internet', 'word_freq_order', 'word_freq_mail', 'word_freq_receive', 'word_freq_will', 'word_freq_people',
               'word_freq_report', 'word_freq_addresses', 'word_freq_free', 'word_freq_business', 'word_freq_email', 'word_freq_you',
               'word_freq_credit', 'word_freq_your', 'word_freq_font', 'word_freq_000', 'word_freq_money', 'word_freq_hp', 'word_freq_hpl',
               'word_freq_george', 'word_freq_650', 'word_freq_lab', 'word_freq_labs', 'word_freq_telnet', 'word_freq_857', 'word_freq_data',
               'word_freq_415', 'word_freq_85', 'word_freq_technology', 'word_freq_1999', 'word_freq_parts', 'word_freq_pm', 'word_freq_direct',
               'word_freq_cs', 'word_freq_meeting', 'word_freq_original', 'word_freq_project', 'word_freq_re', 'word_freq_edu', 'word_freq_table',
               'word_freq_conference', 'char_freq_;', 'char_freq_(', 'char_freq_[', 'char_freq_!', 'char_freq_$', 'char_freq_#',
               'capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total']


    output_columns = ['is_spam']

    all_columns = input_columns + output_columns
    
    data = pd.read_csv(fileName, names=all_columns, index_col=False)
    
    return input_columns,output_columns,all_columns,data
```

```python
input_columns,output_columns,all_columns,data  = load_spam_data('spambase\spambase.data')
```

# 3. Splitting data (Training / Testing / Validation ) 

```python
X = data[input_columns]
Y = data[output_columns]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split

def train_val_test_split(X, y, train_size, val_size, test_size):
    X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size = test_size)
    relative_train_size = train_size / (val_size + train_size)
    X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val,
                                                      train_size = relative_train_size, test_size = 1-relative_train_size)
    return X_train, X_val, X_test, y_train, y_val, y_test
```

```python
X_train, X_val, X_test, y_train, y_val, y_test = train_val_test_split(X, Y, 0.7, 0.1, 0.2)
```

```python
len(X_train), len(X_val), len(X_test), len(y_train), len(y_val), len(y_test)
```

# 4. StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
X_val=ss.fit_transform(X_val)
```

```python
X_train
```

# 5. Creating Neural Network Model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Dropout
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
#Creating a pipeline
model = Sequential()

#1st hidden layer with input layer
model.add(Dense(units=300,activation="relu",input_dim=57))

#2nd hidden layer
model.add(Dense(units=200,activation="relu",))

#3rd hidden layer
model.add(Dense(units=100,activation="relu",))

#output layer
model.add(Dense(units=1,activation="sigmoid"))

```

```python
model.summary()
```

# 6. Compiling Model

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

# 7. Training Model

```python
model_his=model.fit(X_train,y_train, validation_data=(X_val, y_val), batch_size=16,epochs=100,verbose=1)
```

# 8. Evaluating Model

```python
scores = model.evaluate(X_train,y_train, verbose=1)
print(scores)
```

# 9. Testing Model

```python
y_pred=model.predict(X_test)
y_pred = (y_pred > 0.45)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```
